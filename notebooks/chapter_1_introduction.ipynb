```python
import transformers
from transformers import pipeline
```

## document classification

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("Hate-speech-CNERG/dehatebert-mono-english")
model = AutoModelForSequenceClassification.from_pretrained("Hate-speech-CNERG/dehatebert-mono-english")

text = "Kill all the Rohingyas and push them out of the country."

pipe = pipeline("text-classification", model=model, tokenizer=tokenizer)
prediction = pipe(text)
print(prediction)
```

# NER model

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained('ctrlbuzz/bert-addresses')
model = AutoModelForTokenClassification.from_pretrained("ctrlbuzz/bert-addresses")
nlp = pipeline("ner", model=model, tokenizer=tokenizer)
text = "While Maria was representing Johnson & Associates at a conference in Spain, she mailed me a letter from her new office at 123 Elm St., Apt. 4B, Springfield, IL.",

annotations = nlp(text)
print(annotations)
```

## machine translation

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-bn-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-bn-en")

pipe = pipeline('translation', model=model, tokenizer=tokenizer)

text = 'আমি কৃতজ্ঞ কারণ আপনি আমার জন্য অনেক কিছু করেছেন।'
print(pipe(text))
```

## grammar correction

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Unbabel/gec-t5_small")
model = AutoModelForSeq2SeqLM.from_pretrained("Unbabel/gec-t5_small")

pipe = pipeline('text2text-generation', model=model, tokenizer=tokenizer)

sentence = 'i like to swimming'

print(pipe(sentence))
```

## question answering

```python
context = """
    India is the seventh largest country in the world.
    By 2030 India is expected to overtake China as the country with the highest population.
    The national bird of India is the peacock.
    Every day, 5,000 so-called dabbawallas deliver 200,000 lunches to office workers in Mumbai.
    The capital of India is New Delhi.
"""
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering", model="deepset/roberta-base-squad2")

answer = question_answerer(question='What is the national bird of India', context=context)
print(answer)
```

## Summarisation

```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/pegasus-cnn_dailymail")
model = AutoModelForSeq2SeqLM.from_pretrained("google/pegasus-cnn_dailymail")

# Example dialogue for summarization
dialogue = [
    '#Person1#: Happy Birthday, this is for you, Brian.',
    '#Person2#: Im so happy you remember, please come in and enjoy the party. Everyones here, Im sure you have a good time.',
    '#Person1#: Brian, may I have a pleasure to have a dance with you?',
    '#Person2#: Ok.',
    '#Person1#: This is really wonderful party.',
    '#Person2#: Yes, you are always popular with everyone. and you look very pretty today.',
    '#Person1#: Thanks, thats very kind of you to say. I hope my necklace goes with my dress, and they both make me look good I feel.',
    '#Person2#: You look great, you are absolutely glowing.',
    '#Person1#: Thanks, this is a fine party. We should have a drink together to celebrate your birthday'
]
full_dialogue = "\n".join(dialogue)
summarization_pipeline = pipeline("summarization", model=model, tokenizer=tokenizer)
summary = summarization_pipeline(full_dialogue)
print(summary)
```
